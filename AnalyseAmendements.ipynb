```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
location = "http://www.nosdeputes.fr/synthese/data/csv"
```

```python
df = pd.read_csv(location, sep=';')
```

```python
df[df['id'] == 511]
```

```python
df.columns.values
```

```python
df.groupby('groupe_sigle')['groupe_sigle'].count().head(10)
```

```python
def colorByGroupe(groupe):
    if groupe == 'SER'  : return 'pink'
    if groupe == 'LR'   : return 'blue'
    if groupe == 'GDR'  : return 'red'
    if groupe == 'NI'   : return 'yellow'
    if groupe == 'UDI'  : return 'lightblue'
    if groupe == 'RRDP' : return 'black'
    return 'white'
        
        
df['color']=df['groupe_sigle'].map(colorByGroupe)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import plotly
plotly.offline.init_notebook_mode()
```

```python
df_SER = df[df['groupe_sigle']=='SER']
df_LR = df[df['groupe_sigle']=='LR']
df_GDR = df[df['groupe_sigle']=='GDR']
df_UDI = df[df['groupe_sigle']=='UDI']
df_RRDP = df[df['groupe_sigle']=='RRDP']
df_NI = df[df['groupe_sigle']=='NI']

traceSER = go.Scatter(
    x= df_SER['amendements_signes'],
    y= df_SER['amendements_adoptes'],
    name = 'SER',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'pink',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    ),
    text = df_SER['nom']
)
traceLR = go.Scatter(
    x= df_LR['amendements_signes'],
    y= df_LR['amendements_adoptes'],
    name = 'LR',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'blue',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    ),
    text = df_LR['nom']
)
traceUDI = go.Scatter(
    x= df_UDI['amendements_signes'],
    y= df_UDI['amendements_adoptes'],
    name = 'UDI',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'lightblue',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    ),
    text = df_UDI['nom']
)
traceRRDP = go.Scatter(
    x= df_RRDP['amendements_signes'],
    y= df_RRDP['amendements_adoptes'],
    name = 'UDI',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'black',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    ),
    text = df_RRDP['nom']
)
traceGDR = go.Scatter(
    x= df_GDR['amendements_signes'],
    y= df_GDR['amendements_adoptes'],
    name = 'GDR',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'red',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    ),
    text = df_GDR['nom']
)
traceNI = go.Scatter(
    x= df_NI['amendements_signes'],
    y= df_NI['amendements_adoptes'],
    name = 'NI',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'yellow',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    ),
    text = df_NI['nom']
)

data = [traceSER, traceLR, traceUDI, traceRRDP, traceGDR, traceNI]

layout= go.Layout(
    title= 'Amendements signés vs amendements adoptés',
    hovermode= 'closest',
    xaxis= dict(
        title= 'Amendement signés',
        ticklen= 5,
        zeroline= False,
        gridwidth= 2,
    ),
    yaxis=dict(
        title= 'Amendements adoptés',
        ticklen= 5,
        gridwidth= 2,
        range= [-100,875]
    ),
    showlegend= False
)


fig = dict(data=data, layout=layout)
plotly.offline.iplot(fig, filename='styled-scatter')
```
